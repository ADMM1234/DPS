```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Cleavland.csv',names=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num'])
```

```python
plt.hist(df['chol'],color='pink')
```

```python
sns.stripplot(data=df,x='sex',y='trestbps')
```

```python
sns.swarmplot(data=df, x='sex', y='thalach')
```
