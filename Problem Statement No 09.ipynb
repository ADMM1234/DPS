Perform the following operations using Python on the Facebook metrics data sets

a. Create data subsets for type of post

b. Merge two subsets

c. Sort Data on Page total likes

d. Transposing Data

e. Melting Data to long format

f. Casting data to wide format

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dataset_facebook.csv',sep =';')
```

```python
df.head()
```

```python
df['Type'].unique()
```

### a. Create data subsets for type of post

```python
Photo=df[df['Type']=='Photo']
Status=df[df['Type']=='Status']
Link=df[df['Type']=='Link']
Video=df[df['Type']=='Video']
```

```python
Photo.head()
```

```python
Status.head()
```

```python
Link.head()
```

```python
Video.head()
```

### b. Merge two subsets

```python
Merge=[Photo,Status]
Merged_Photo_Status=pd.concat(Merge)
```

```python
Merged_Photo_Status.head()
```

```python
Merged_Photo_Status['Type'].unique()
```

### c. Sort Data on Page total likes

```python
df.sort_values(by='Page total likes',ascending=False)
```

```python
df.sort_values(by='Page total likes',ascending=True)
```

### d. Transposing Data

```python
Transposed_Data=df.transpose()
```

```python
Transposed_Data.head()
```

### e. Melting Data to long format

```python
df_To_Use = df[['Type','Page total likes','Category']]
```

```python
df_To_Use.head()
```

```python
df_melt = pd.melt(df_To_Use,id_vars=['Type'],value_vars=['Page total likes','Category'])
```

```python
df_melt.head()
```

```python
df_melt[df_melt['variable']=='Category'].head()
```

### f. Casting data to wide format

```python
df_To_Cast1=df_To_Use[df_To_Use['Type']=='Photo'].head(1)
df_To_Cast2=df_To_Use[df_To_Use['Type']=='Video'].head(1)
df_To_Cast3=df_To_Use[df_To_Use['Type']=='Link'].head(1)
df_To_Cast4=df_To_Use[df_To_Use['Type']=='Status'].head(1)
```

```python
df_To_Cast=pd.concat([df_To_Cast1,df_To_Cast2,df_To_Cast3,df_To_Cast4])
```

```python
df_To_Cast.head()
```

```python
df_To_Cast.pivot(index='Type',columns='Category',values='Page total likes')
```
