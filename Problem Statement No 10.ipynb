Perform the following operations using Python on the Iris data sets

g. Create data subsets for different species

h. Merge two subsets

i. Sort Data Petal Length

j. Transposing Data

k. Melting Data to long format

l. Casting data to wide format

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Iris.csv',names=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm','class'])
```

```python
df.head()
```

```python
df['class'].unique()
```

### g. Create data subsets for different species

```python
Iris_setosa=df[df['class']=='Iris-setosa']
Iris_versicolor=df[df['class']=='Iris-versicolor']
Iris_virginica=df[df['class']=='Iris-virginica']
```

```python
Iris_setosa.head()
```

```python
Iris_versicolor.head()
```

```python
Iris_virginica.head()
```

### h. Merge two subsets

```python
Merge=[Iris_versicolor,Iris_virginica]
Merged_versicolor_virginica=pd.concat(Merge)
```

```python
Merged_versicolor_virginica.head()
```

```python
Merged_versicolor_virginica['class'].unique()
```

### i. Sort Data Petal Length

```python
df.info()
```

```python
df['petal length in cm']=df['petal length in cm'].astype('float')
```

```python
df.sort_values(by='petal length in cm',ascending=True)
```

```python
df.sort_values(by='petal length in cm',ascending=False)
```

### j. Transposing Data

```python
Transposed_Data=df.transpose()
```

```python
Transposed_Data.head()
```

### k. Melting Data to long format

```python
df_melt = pd.melt(df,id_vars=['class'],value_vars=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
df_melt.head()
```

```python
df_melt[df_melt['variable']=='sepal width in cm'].head()
```

```python
df_melt[df_melt['variable']=='petal length in cm'].head()
```

```python
df_melt[df_melt['variable']=='petal width in cm'].head()
```

### l. Casting data to wide format

```python
df.index
```

```python
df_To_Cast=df.drop(['sepal width in cm','petal length in cm','petal width in cm'],axis='columns')
```

```python
df_To_Cast.info()
```

```python
df_To_Cast['index_col'] = df_To_Cast.index
```

```python
df_To_Cast.pivot(index='index_col',columns='class',values='sepal length in cm')
```
